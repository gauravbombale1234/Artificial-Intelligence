```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv('cancer_classification.csv')
```

```python
df.info()
```

### Train Test Split

```python
X=df.drop('benign_0__mal_1',axis=1).values
y=df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=125)
```

### Scaling Data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

### Creating Model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
```

```python
X_train.shape
```

```python
model=Sequential()

model.add(Dense(units=30,activation='relu'))

model.add(Dense(units=15,activation="relu"))

model.add(Dense(units=1,activation='sigmoid'))

#for  a binary classification problem
model.compile(loss="binary_crossentropy",optimizer='adam')
```

### Training the model

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test, y_test),verbose=1)
```

```python
model_loss=pd.DataFrame(model.history.history)
```

```python
#Model Loss
```

```python
model_loss.plot()
```

```python
#Inference of Graph: Initially there is no loss but gradually loss is increasing.
```

### Example Two : Early Stopping

```python
#we obviously trained too much! Let's use early stopping to track the val_loss and stop training once it begins increasing 
#too much!
```

```python
model=Sequential()
model.add(Dense(units=30,activation='relu'))
model.add(Dense(units=15,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop=EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=0)
#If you are working on accuracy mode='Max', if loss then mode='min'
#patience:Number of epochs with no improvement after which training will beDefaults to 0
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),verbose=1,callbacks=[early_stop])
```

```python
model_loss=pd.DataFrame(model.history.history)
model_loss.plot()
```

### Example Three: Adding in Dropout Layers

```python
from tensorflow.keras.layers import Dropout
```

```python
model1=Sequential()
model1.add(Dense(units=30,activation='relu'))
model1.add(Dropout(0.5))
#it is between 0 to 100% , here we are taking  50%
model1.add(Dense(units=15,activation='relu'))
model1.add(Dropout(0.5))
#it is between 0 to 100% , here we are taking  50%
model1.add(Dense(units=1,activation='sigmoid'))
model1.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model1.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),verbose=1,callbacks=[early_stop])
```

```python
model_loss=pd.DataFrame(model1.history.history)
model_loss.plot()
```

```python
predictions=(model.predict(X_test) > 0.5).astype('int32')
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
