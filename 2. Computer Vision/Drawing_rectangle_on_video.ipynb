```python
import cv2

cap=cv2.VideoCapture(0)

#automatically grab width and height of the video feed
#(return float wich we need to convert to integer for later on)
width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

#we using // here because in python // allows for int classical diiviion
#because we canit pass a float to the cv2.reactangle function

#coordinartes for Rectangle, centre point rectangle
x=width//2
y=height//2

#width and height adding 1/4 of width and height from the center
w=width//4
h=height//4

while True:
    #capture frame-by-frame
    ret, frame=cap.read()
    
    #draw a recatngle on stream
    cv2.rectangle(frame,(x,y), (x+w, y+h), color=(0,0,255), thickness=4)
    
    #display the resulting framme
    cv2.imshow('frame', frame)
    
    #this command lets us quit te "q" button on a keyboard
    #simply pressing X on the button window wont work
    
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
        
#when everything is done release the capture

cap.release()
cv2.destroyAllWindows()
```

### Interactive frame 

```python
import cv2

# Define global variables
pt1 = (0, 0)
pt2 = (0, 0)
topLeft_clicked = False
botRight_clicked = False

'''
Following loop continuously captures frames from the webcam. 
If topLeft_clicked is True, a small red circle is drawn at pt1. If both topLeft_clicked and 
botRight_clicked are True, a red rectangle is drawn between pt1 and pt2. cv2.imshow('Test', frame) 
displays the video frame in the "Test" window. cv2.waitKey(1) & 0xFF == ord('q') checks for the 'q'
key press to exit the loop and close the window.
'''

# Mouse callback function to draw a rectangle
def draw_rectangle(event, x, y, flags, param):
    global pt1, pt2, topLeft_clicked, botRight_clicked
    
    # Check if the left mouse button was clicked
    if event == cv2.EVENT_LBUTTONDOWN:
        
        # Reset the rectangle if both corners are already clicked
        if topLeft_clicked and botRight_clicked:
            topLeft_clicked = False
            botRight_clicked = False
            pt1 = (0, 0)
            pt2 = (0, 0)
        
        # Capture the first corner
        if not topLeft_clicked:
            pt1 = (x, y)
            topLeft_clicked = True
            
        # Capture the second corner
        elif not botRight_clicked:
            pt2 = (x, y)
            botRight_clicked = True

# Start capturing video
cap = cv2.VideoCapture(0)

# Create a named window for the video stream
cv2.namedWindow('Test')

# Bind the draw_rectangle function to mouse clicks
cv2.setMouseCallback('Test', draw_rectangle)

# Main loop to display the video stream and handle drawing
while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    
    # Draw a circle on the first corner if it's clicked
    if topLeft_clicked:
        cv2.circle(frame, center=pt1, radius=5, color=(0, 0, 255), thickness=-1)
    
    # Draw the rectangle if both corners are clicked
    if topLeft_clicked and botRight_clicked:
        cv2.rectangle(frame, pt1, pt2, (0, 0, 255), 2)
    
    # Display the resulting frame
    cv2.imshow('Test', frame)
    
    # Press 'q' to quit the loop and close the window
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
# When everything is done, release the capture and close windows
cap.release()
cv2.destroyAllWindows()
```
