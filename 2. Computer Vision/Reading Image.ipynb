### By ratio

```python
import cv2

w_ratio=0.5  # Scaling factor for width
h_ratio=0.5   # Scaling factor for height

# Resize the image
new_img=cv2.resize(img_rgb,(0,0),fx=w_ratio,fy=h_ratio)
plt.imshow(new_img)

'''
cv2.resize:
This is a function from the opencv library that resizes th image to a specified size

img_rgb:
This is the input image that you want to resize. It is assumed to be in RGB format.

(0, 0):
This is the desired size of the output image. Setting both dimensions to e means that the size will be determined
by the scaling factors provided.
'''
```

### Flipping Image

```python
%matplotlib inline
```

```python
# Along central x axis
new_img=cv2.flip(new_img,0)
plt.imshow(new_img)
'''
This is the flip code that specifies how the image should be flipped:
        O means flipping around the x-axis (vertical flip).
        1 means flipping around the y-axis (horizontal flip).
       -1 means flipping around both axes (both vertical and horizontal flip
'''
```

```python
# Along central y axis
new_img=cv2.flip(new_img,1)
plt.imshow(new_img)
```

```python
# Along central both axis
new_img=cv2.flip(new_img,-1)
plt.imshow(new_img)
```
