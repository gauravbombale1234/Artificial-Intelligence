```python
'''
Blending and Pasting Images

For some computer vision systems, we'll want to be able to post our own image on top of an 
already existing image or video. We may also want to blend images, maybe we want to have a 
"highlight" effect instead of just a solid box or empty rectangle.
'''
```

```python
import cv2
```

```python
# The images
img1=cv2.imread("dog_backpack.jpg")
img2=cv2.imread("watermark_no_copy.png")
```

```python
img1.shape
```

```python
img2.shape
```

```python
plt.imshow(img1)
```

```python
img1=cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2=cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
img1=cv2.resize(img1,(1200,1200))
img2=cv2.resize(img2,(1200,1200))
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
### Blending the image
blended = cv2.addWeighted(src1=img1, alpha=0.7, src2=img2, beta=0.3, gamma=0)
# Try by changing the value of alpha and beta
```

```python
plt.imshow(blended)
```

### Different Threshold Type

 #####  Binary

```python
ret,thresh1=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
ret
```

```python
plt.imshow(thresh1,cmap='gray')
```

#### Binary Inverse

```python
ret,thresh2=cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
plt.imshow(thresh2,cmap='gray')
```

#### Threshold Truncation

```python
ret,thresh3=cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
plt.imshow(thresh3,cmap='gray')
```

#### Threshold to Zero

```python
ret,thresh4=cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
plt.imshow(thresh4,cmap='gray')
```
